```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stat
```

```python
nut = pd.read_csv('nutrients.csv')
oxy = pd.read_csv('oxygen.csv')
```

```python
nut['conc (uM)'][2]=nut['conc (uM)'][2]/10000 #forgot to add decimal to this data point.. oops!
nut
```

```python
nutdepth0=[[],[],[]]
nutdepth5=[[],[],[]]

for i in range(len(nut)):
    if nut['depth'][i] == 0:
        nutdepth0[nut['station'][i]-1].append(nut['conc (uM)'][i])
    if nut['depth'][i] == 5:
        nutdepth5[nut['station'][i]-1].append(nut['conc (uM)'][i])
```

```python
for i in range(len(nutdepth0)): nutdepth0[i]=stat.mean(nutdepth0[i]) 
for i in range(len(nutdepth5)): nutdepth5[i]=stat.mean(nutdepth5[i]) 
```

```python
stations = ['station 1', 'station 2', 'station 3']

plt.plot(stations,nutdepth0, label='0m depth samples')
plt.plot(stations,nutdepth5, label='5m depth samples')
plt.ylim(0,20)
plt.legend()
plt.ylabel('Nitrate Concentration [uM]')
```

```python
oxdepth0=[[],[],[]]
oxdepth5=[[],[],[]]

for i in range(len(oxy)):
    if oxy['depth'][i] == 0:
        oxdepth0[oxy['station'][i]-1].append(oxy['ml/l O2'][i])
    if oxy['depth'][i] == 5:
        oxdepth5[oxy['station'][i]-1].append(oxy['ml/l O2'][i])
        
for i in range(len(oxdepth0)): oxdepth0[i]=stat.mean(oxdepth0[i]) 
for i in range(len(oxdepth5)): oxdepth5[i]=stat.mean(oxdepth5[i]) 
```

```python
plt.plot(stations,oxdepth0, label='0m depth samples')
plt.plot(stations,oxdepth5, label='5m depth samples')
plt.ylim(0,10)
plt.ylabel('Dissolved Oxygen [mL/L O2]')
plt.legend()
```
